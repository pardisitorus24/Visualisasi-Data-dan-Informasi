### Jawab:

Mari kita gunakan Plotly untuk membuat visualisasi interaktif. Saya akan membuat beberapa visualisasi yang lebih rinci, seperti:

1. Histogram interaktif dari distribusi jumlah streaming berdasarkan tahun rilis.
2. Bubble chart interaktif yang menunjukkan hubungan antara jumlah streaming dan karakteristik audio (danceability, energy, valence).

```python
import pandas as pd
import plotly.express as px

# Memuat data dari jalur file CSV yang benar
file_path = '/content/Spotify Most Streamed Songs.csv'
data = pd.read_csv(file_path)

# Convert 'streams' column to numeric, handling errors by coercing to NaN
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')

# 1. Histogram interaktif dari distribusi jumlah streaming berdasarkan tahun rilis
fig1 = px.histogram(data, x='released_year', y='streams', title='Distribusi Jumlah Streaming Berdasarkan Tahun Rilis',
                   labels={'released_year': 'Tahun Rilis', 'streams': 'Jumlah Streaming'},
                   color='released_year',
                   hover_data=['track_name', 'artist(s)_name'],
                   nbins=15, height=400)
fig1.show()

# 2. Bubble chart interaktif untuk melihat hubungan antara danceability, energy, valence, dan jumlah streaming
# Filter out NaN values from 'streams' column before creating the scatter plot
filtered_data = data.dropna(subset=['streams'])  # Drop rows with NaN in 'streams'

fig2 = px.scatter(filtered_data, x='danceability_%', y='streams',  # Use filtered_data
                  size='streams', color='valence_%',
                  hover_name='track_name', hover_data=['artist(s)_name'],
                  title='Perbandingan Danceability, Energy, Valence dengan Jumlah Streaming',
                  labels={'danceability_%': 'Danceability (%)', 'streams': 'Jumlah Streaming', 'valence_%': 'Valence (%)'},
                  size_max=60)
fig2.show()
```
