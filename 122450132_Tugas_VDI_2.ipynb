## Visualisasi Data yang Baik

Visualisasi Data yang Baik.png

* ## **Penjelasan:**
    * **Jelas:** Grafik ini dengan jelas menunjukkan aliran migrasi global dari tahun 1990 hingga 2017. Penggunaan warna yang berbeda untuk setiap benua memudahkan identifikasi asal dan tujuan migrasi.
    * **Akurat:** Data yang disajikan akurat dan berasal dari sumber yang terpercaya (PBB).
    * **Menarik:** Desainnya bersih dan mudah dibaca, dengan penggunaan animasi yang halus untuk menunjukkan perubahan migrasi dari waktu ke waktu.
    * **Relevan:** Visualisasi ini sangat relevan dengan isu migrasi global yang sedang hangat diperbincangkan. Ini membantu pembaca memahami skala dan kompleksitas migrasi manusia di seluruh dunia.

## Visualisasi Data yang Buruk

Visualisasi Data yang Buruk.png

* ## **Penjelasan:**
    * **Tidak Jelas:** Efek 3D membuat sulit untuk membandingkan ukuran irisan secara akurat. Irisan yang berada di depan tampak lebih besar daripada yang sebenarnya.
    * **Tidak Akurat:** Distorsi visual yang disebabkan oleh efek 3D dapat menyebabkan pembaca salah menginterpretasikan proporsi data.
    * **Tidak Menarik:** Desain 3D seringkali terlihat berlebihan dan tidak elegan. Selain itu, penggunaan warna yang tidak harmonis dapat membuat grafik sulit dibaca.
    * **Tidak Relevan:** Penggunaan diagram lingkaran 3D seringkali tidak perlu dan tidak menambah nilai informasi pada visualisasi. Ada bentuk visualisasi lain yang lebih efektif untuk menampilkan proporsi, seperti diagram batang atau diagram donat.
